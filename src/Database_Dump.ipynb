```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/Project/Cardio_Cleaned.csv")
display(df.head())
display(df.info())
```

```python
import sqlite3

conn = sqlite3.connect('cardio_database.db')
cursor = conn.cursor()

print("Connected to SQLite database.")
```

```python
df.to_sql('cardio_data', conn, if_exists='replace', index=False)

print("Data successfully dumped to 'cardio_data' table.")
```

```python
cursor.execute("SELECT * FROM cardio_data")
rows = cursor.fetchall()

for row in rows:
    print(row)
```
